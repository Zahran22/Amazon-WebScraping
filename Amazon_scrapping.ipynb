```python
from selenium import webdriver


from selenium.webdriver.common.by import By
import time

# Set up the webdriver with webdriver-manager

driver = webdriver.Chrome()

# Open the Amazon website
driver.get("https://www.amazon.eg/s?k=samsung&rh=p_89%3Asamsung&language=en&ref=SQEG-WEB-SR301")

# Wait for the page to load
time.sleep(3)

# Find all product elements
products = driver.find_elements(By.XPATH, '//div[contains(@class, "s-result-item")]')

for product in products:
    try:
        # Get the title
        title_element = product.find_element(By.XPATH, 'xpath',"//span[@class='a-size-base-plus a-color-base a-text-normal']")
        title = title_element.text

        # Get the number of ratings
        ratings_element = product.find_element(By.XPATH, './/span[@class="a-size-base"]')
        ratings = ratings_element.text

        # Get the rating
        rating_element = product.find_element(By.XPATH, './/span[@class="a-icon-alt"]')
        rating = rating_element.get_attribute("innerHTML")

        # Get the price
        try:
            price_element = product.find_element(By.XPATH, './/span[@class="a-price-whole"]')
            price = price_element.text
        except:
            price = "N/A"

        # Get the image URL
        image_element = product.find_element(By.XPATH, './/img[@class="s-image"]')
        image_url = image_element.get_attribute("src")

        print(f"Title: {title}")
        print(f"Number of Ratings: {ratings}")
        print(f"Rating: {rating}")
        print(f"Price: {price}")
        print(f"Image URL: {image_url}")
        print("-" * 40)

    except Exception as e:
        print(f"Error: {e}")
        continue

# Close the browser
driver.quit()

```
